```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Valid intervals: [1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo]
yf.download("TATAMOTORS.NS",interval="1d")
```

```python
asset_data= yf.download("TATAMOTORS.NS",start="2021-01-01",end="2022-01-01")
asset_data
```

```python
max_value = asset_data['Close'].max()
max_value
```

```python
asset_data[asset_data['Close']==asset_data['Close'].max()]
```

```python
min_value = asset_data['Close'].min()
min_value
```

```python
asset_data[asset_data['Close']==asset_data['Close'].min()]
```

```python
asset_data['Close'].describe()
```

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(30, 15))
plt.title(' TATA MOTORS 2021-01-01 to 2021-01-01')
plt.plot(asset_data.index,asset_data['Close'],color='red',linewidth=3)
plt.xlabel('Date')
plt.ylabel('Value')
plt.margins(0.01)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(30, 15))
plt.title(' TATA MOTORS 2021-01-01 to 2021-01-01')
plt.plot(asset_data.index,asset_data['Open'],color='Green',linewidth=3)
plt.xlabel('Date')
plt.ylabel('Value')
plt.margins(0.01)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(30, 15))
plt.title(' TATA MOTORS 2021-01-01 to 2021-01-01')
plt.plot(asset_data.index,asset_data['Open'],color='Green',linewidth=3,label='Open Price')
plt.plot(asset_data.index,asset_data['Close'],color='red',linewidth=3,label='Close Price')
plt.xlabel('Date')
plt.ylabel('Value')
plt.margins(0.01)
plt.legend()
plt.grid()
plt.show()
```

```python
asset_data['profit_loss']=asset_data['Open']-asset_data['Close']
```

```python
daily_returns = asset_data['Adj Close'].pct_change()
monthly_returns = asset_data['Adj Close'].resample('M').ffill().pct_change()
```

```python
plt.figure(figsize=(30, 15))
plt.title(' TATA MOTORS 2021-01-01 to 2021-01-01 Daily Returns')
plt.plot(daily_returns,color='Green',linewidth=3)
plt.margins(0.01)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(30, 15))
plt.title(' TATA MOTORS 2021-01-01 to 2021-01-01 Montly Returns')
plt.plot(monthly_returns,color='Green',linewidth=3)
plt.margins(0.01)
plt.grid()
plt.show()
```
